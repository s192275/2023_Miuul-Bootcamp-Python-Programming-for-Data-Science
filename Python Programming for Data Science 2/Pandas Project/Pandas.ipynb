```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
df.groupby('time')['total_bill'].agg(['min','max','mean'])
```

```python
df.groupby(['day','time'])['total_bill'].agg(['min','max','mean'])
```

```python
df.dtypes
#df['size'] = df['size'].astype('float')
```

```python
selected_data = df.loc[(df['time'] == 'Lunch') & (df['sex'] == 'Female')]
total_by_day = selected_data.groupby('day')['total_bill', 'tip'].sum()
min_by_day = selected_data.groupby('day')['total_bill', 'tip'].min()
max_by_day = selected_data.groupby('day')['total_bill', 'tip'].max()
mean_by_day = selected_data.groupby('day')['total_bill', 'tip'].mean()

print("Toplam (Total) ve Bahşiş (Tip) değerlerinin gün (day) bilgisine göre toplamı:")
print(total_by_day)

print("\nToplam (Total) ve Bahşiş (Tip) değerlerinin gün (day) bilgisine göre minimumu:")
print(min_by_day)

print("\nToplam (Total) ve Bahşiş (Tip) değerlerinin gün (day) bilgisine göre maksimumu:")
print(max_by_day)

print("\nToplam (Total) ve Bahşiş (Tip) değerlerinin gün (day) bilgisine göre ortalaması:")
print(mean_by_day)
```
